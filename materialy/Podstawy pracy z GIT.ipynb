# Systemy Kontroli Wersji

## Scentralizowane

(Centralized Version Control System - CVCS)

Przykłady: SVN (Subversion), TFS, CSV

Główny serwer przechowuje całe repozytorium. Lokalnie przechowywany jest jeden commit (jedna, aktualnie rozwijana wersja)

### zalety

- może być wydajniejszy przy dużych projektach z dużymi plikami (szczególnie binarnymi). Przechowywanie pełnej histori zmian lokalnie i wysyłanie tych zmian poprzez sieć moze być problematyczne.

### wady

- tylko jedna maszyna przechowuje całe repozytorium. W przypadku awarii serwera czy problemów sieciowych praca może być utrudniona lub nawet niemożliwa

- dość niska wydajność w większości zastosowań. Lokalnie przechowywany jest tylko jeden commit. Każda zmiana wymaga komunikacji sieciowej. Nawet w dobie szybkiego internetu operacje sieciowe będą wolniejsze niż przeprowadzane lokalnie.



## Zdecentralizowane (rozproszone)

(Distributed Version Control System - DVCS)

Przykłady: Git, Mercurial (HG), Bazaar

Każda maszyna może przechowywać całe repozytorium. 

###  Zalety

- zmiany można zapisywać lokalnie (offline) i co jakiś czas je synchroniziować z zewnętrznymi repozytoriami

- lepsz wydajność (zmiany lokalne)

- różne sposoby pracy (workflows)

- Możliwość uporządkowania, zmiany rewizji (commitow). Zanim 


### Wady

- dość słaba obsługa dużych plików

# Działanie systemu wersjonowania git

## Commit 

Element wspólny wszystkich systemów wersjonowania.

Inne nazwy to: changeset, revision, version

W przypadku git jest to stan plików śledzonych w repozytorium w danym momencie czasu. Commit oznaczony jest specjalnym, wyliczalnym ciągkiem znaków - SHA1

Commit oprócz stanu zawiera informacje o autorze, czasie utworzenia oraz opis zmian (commit message)

## Branch

Każdy projekt zawiera przynajmniej jedną gałąź. Gałęzi możę być jednak wiele. Można wyodrębniać nowe gałęzie, można je scalać 

## Projekt

Projekt w git zawiera historię zmian zapisaną w commitach oraz gałęziach (branches)

## Submoduły

W ramach repozytorium możemy mieć też foldery zawierające inne repozytoria. W języku gita nazywa się to submodułem.
Operowanie na tych repozytoriach w zasadzie prowadzone jest niezależenie. 
Główny moduł zawiera informację o tym jaka obecnie rewizja obowiązuje
