### **Zadanie: Prosty edytor tekstu z filtrami, szyfrowaniem i deszyfrowaniem**

#### **Opis zadania**
Twoim zadaniem jest stworzenie prostego edytora tekstu, który pozwala użytkownikowi przetwarzać tekst za pomocą różnych filtrów. Program powinien obsługiwać predefiniowane filtry, takie jak konwersja tekstu do wielkich liter, usuwanie spacji czy odwracanie tekstu. Dodatkowo, użytkownik może skorzystać z funkcji szyfrowania i deszyfrowania tekstu za pomocą **szyfru Cezara**. Użytkownik ma także możliwość dynamicznego dodawania własnych filtrów w czasie działania programu.

---

#### **Wymagania funkcjonalne**
1. **Dostępne filtry**:
   - Konwersja do wielkich liter.
   - Usuwanie spacji.
   - Odwrócenie tekstu.
   - Wyświetlanie liczby słów.
   - **Szyfrowanie (szyfr Cezara)** — przesuwa litery w alfabecie o podany klucz.
   - **Deszyfrowanie (szyfr Cezara)** — odwrotność szyfrowania z użyciem klucza.

2. **Dynamiczne dodawanie filtrów**:
   - Użytkownik może wprowadzić nazwę nowego filtra oraz jego definicję w postaci funkcji lambda.

3. **Interfejs użytkownika**:
   - Program działa w trybie tekstowym, umożliwiając wybór różnych opcji:
     - Wprowadzenie tekstu do edycji.
     - Wybór filtra do zastosowania.
     - Dodanie nowego filtra.
     - Wyświetlenie dostępnych filtrów.
     - Wyjście z programu.

4. **Bezpieczeństwo danych**:
   - Obsłuż nieprawidłowe dane wejściowe (np. brak filtra o podanej nazwie, niepoprawny klucz szyfrowania).

---

#### **Przykład działania programu**

```
Podaj tekst do edycji:
> Ala ma kota

Wybierz opcję:
1. Zastosuj filtr
2. Dodaj nowy filtr
3. Wyświetl dostępne filtry
4. Wyjdź
> 1

Dostępne filtry:
1. Konwersja do wielkich liter
2. Usuwanie spacji
3. Odwrócenie tekstu
4. Wyświetlanie liczby słów
5. Szyfrowanie (Cezar)
6. Deszyfrowanie (Cezar)

Wybierz filtr:
> 5
Podaj klucz szyfrowania (liczba całkowita):
> 3
Wynik: Dod pd nrwd

Wybierz opcję:
> 1

Wybierz filtr:
> 6
Podaj klucz deszyfrowania (liczba całkowita):
> 3
Wynik: Ala ma kota

Wybierz opcję:
> 2

Podaj nazwę nowego filtra:
> Zamiana na małe litery
Podaj definicję filtra jako lambda (np. lambda text: text.lower()):
> lambda text: text.lower()
Nowy filtr 'Zamiana na małe litery' został dodany!

Wybierz opcję:
> 1

Wybierz filtr:
> 7
Wynik: ala ma kota
```
